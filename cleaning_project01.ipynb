```python
with open("final.txt","r") as f:
    data=f.read()
```

```python
data1 = data.split("\n")
```

```python
data1.pop(0)
records=[]
for line in data1:
    chunks = line.split(",")
    records.append(chunks)
```

```python
def cleaning(records):
    clean_data = []
    for chunk in records:
        if chunk[0] == " ":
            continue
        else:
            user_id = int(chunk[0])
        if chunk[1] in ["unknown"," ","fake user"]:
             continue
        else:
            user_name = chunk[1]
        if chunk[2].strip() in ["?","","twenty-two"]:
            continue
        else:
            age = int(chunk[2])
        if chunk[3] in ["Nowhere"," ","unknown"]:
            continue
        else:
             city = chunk[3]
        if chunk[4] in ["unknown","invalid"]:
            continue
        else:
            signup_date = chunk[4]
        if chunk[5] in ["unknown","NA","invalid_date","none"]:
            continue
        else:
            last_login = chunk[5]
        if chunk[6].strip() in ["?",""]:
            continue
        else:
            amount_spent = int(chunk[6])
        if chunk[7] in ["unknown"]:
            continue
        else:
            status = chunk[7]
        clean_data.append({"user_id":user_id,"user_name":user_name,"age":age,"city":city,"signup_date":signup_date,"last_login":last_login,"amount_spent":amount_spent,"status":status})
    return clean_data
cleaning(records)
```
